```python
from tabulate import tabulate
from decimal import Decimal
import pandas as pd
```

```python
lexical_diversity = {}
for genre in brown.categories():
    type = len(set(brown.words(categories=genre)))
    token = len(list(brown.words(categories=genre)))
    lexical_diversity[genre] = token / type
lexical_diversity = {k: v for k, v in sorted(lexical_diversity.items(), key=lambda item: item[1], reverse=True)}
headers = ["Genre", "Lexical Diversity"]
df = pd.DataFrame([lexical_diversity])
print(tabulate(df.T, headers=headers))
```

```python
import random
```

```python
def get_random_string(length):
    # choose from all lowercase letter
    letters = "abcdefg "
    result_str = ''.join(random.choice(letters) for i in range(length))
    return result_str
    
sentense = get_random_string(100000)
words_list = sentense.split()

fdist = dict(nltk.FreqDist(words_list))
fdist = {k: v for k, v in sorted(fdist.items(), key=lambda item: item[1], reverse=True)}
fdist_list = list(fdist.values())
p = np.arange(1, len(fdist_list)+1)
pl.loglog(p, fdist_list, 'b-')
pl.show()
```

Zipf's law and BA networks follow the power law and count as a scale-free network. Hence, you can see the analytical log graphs look pretty similar. We could compare the two theoretical functions that could see obviously: <br><br> **Zipt's law:**
$$
f = cr^{-s}
$$
Where the $f$ is frequency, the $r$ is the rank, where s and c are parameters that depend on the language and the text. <br><br>
**BA-Networks**
$$
P(k) \sim k^{- \gamma}
$$
Where $P$ is possibility, $k$ is the degrees of that one node, $ \gamma $ depends on the networks but usually will be in the interval of 2 and 3.
